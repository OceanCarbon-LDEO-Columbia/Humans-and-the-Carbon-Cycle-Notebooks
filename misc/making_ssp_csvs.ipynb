## TEMPERATURES

```python
### getting temperatures from dataframe ###

# isolating surface temperature row
rows_to_skip = [*range(1,18,1)]
rows_to_skip.remove(17) # rows that aren't surface temperature

ssp126_rows_to_skip = [*range(1,18,1)]
ssp126_rows_to_skip.remove(7)

temp119 = pd.read_csv('/home/jovyan/galen_course/SSP119_magicc_202408131930.csv',skiprows=rows_to_skip)
temp126 = pd.read_csv('/home/jovyan/galen_course/SSP126_magicc_202408131930.csv',skiprows=ssp126_rows_to_skip)
temp245 = pd.read_csv('/home/jovyan/galen_course/SSP245_magicc_202408131930.csv',skiprows=rows_to_skip)
temp370 = pd.read_csv('/home/jovyan/galen_course/SSP370_magicc_202408131929.csv',skiprows=rows_to_skip)
temp434 = pd.read_csv('/home/jovyan/galen_course/SSP434-over_magicc_202408131929.csv',skiprows=rows_to_skip)
temp585 = pd.read_csv('/home/jovyan/galen_course/SSP585_magicc_202408131252.csv',skiprows=rows_to_skip)

# selecting columns with years and data
years = [*range(1995,2101,1)]
temps119,temps126,temps245,temps370,temps434,temps585 = [],[],[],[],[],[]
for yr in years:
    temps119.append(temp119[str(yr)].values[0]) # temperature values in Kelvin
    temps126.append(temp126[str(yr)].values[0])
    temps245.append(temp245[str(yr)].values[0])
    temps370.append(temp370[str(yr)].values[0])
    temps434.append(temp434[str(yr)].values[0])
    temps585.append(temp585[str(yr)].values[0])
```

```python
scenario_temps = pd.DataFrame(
    data=[temps119,temps126,temps245,temps370,temps434,temps585],
    index=['ssp119','ssp126','ssp245','ssp370','ssp434-over','ssp585'],
    columns=years)
```
